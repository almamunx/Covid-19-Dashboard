```
import bs4
import csv
import requests
import pandas as pd
```

```
site = requests.get('https://worldometers.info/coronavirus')
soup = bs4.BeautifulSoup(site.content,'html.parser')
table = soup.findAll("table",{"id":"main_table_countries_today"})[0]
lists = table.findAll("tr")
```

```
with open("/content/gdrive/My Drive/Colab Notebooks/Covid-19/covid_update.csv", "wt+",encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    for list in lists:
        csv_row = []
        for cell in list.findAll(["td", "th"]):
            csv_row.append(cell.get_text())
        writer.writerow(csv_row)
```
